<ul style="list-style: none; padding-left: 0; margin-left: 0;">
    <li style="display: flex; align-items: center; margin-bottom: 10px;">

### Define: 
- It uses control over the money supply by a nation's central bank to stimulate the economy
   - During a recession / a period of slow growth
   - to contract an economy to fight inflation 
      - **Note**:
        - *Usually use in conjunction with Fiscal Policy*
  
### Expansionary Monetary Policy:
- Recession (slow growth)
   - Central Bank loosens Monetary Policy
     - Money Supply increases
     - Interest Rates decrease
        - Stimulate Investment ---> GPD up!
        - indirectly increases net exports (currency gets weakened: Exports up, Imports down) -->GDP up!

### Contractionary Monetary Policy:
- Inflation (rapid growth)
   - Central Bank tighens Monetary Policy
     - Money Supply decreases
     - Interest Rates increase
        - Discourage Investment ---> GPD down!
        - indirectly decreases net exports (currency gets strengthed: Exports down, Imports up) -->GDP down!

<strong>Define</strong>:<br>
<span style="color: blue; font-style: italic; margin: 0; padding: 0; line-height: 0; font-size: 12pt;">the policy that allows a nation to exchange of goods and services internationally.</span><br>

**Question**
- *How freely a nation should allow imports into its markets?*
  - Protectionist Countries:
     - Erect very high tariffs / **currency manipulation** to imports
  - Others (less concerned):
     - lower trade barriers to encourage trade with others

**Fair vs Unfair Trade**

- Free/Fair Trade:
    - Increased growth
      - Benefited from trade for all ==> better off for all

- UnFair Trade:
    - Countries manipulated with illeal export subsidies, currency...etc.    
      - Results in a "Zero-sum" game where manipulated parties benefited at the expense of others

<strong>Define</strong>:<br>
<span style="color: blue; font-style: italic; margin: 0; padding: 0; line-height: 0; font-size: 12pt;">designed to protect the public and promote the public good. Yet, every regulation entails significant costs on taxpayers, businesses and the broader economy</span><br>


**Areas:**
- Envrionmental Protection
- Employmenet and Labor
- Food & Drug Safety
- Privacy
- Health Care
- Antitrust
- Truthful Advertising

**Supply-side Economics**
- Streamline or eliminate regulations
  - Reduce Business Costs
     - Stimulate Economic Growth
        - Note:
           - Political rhetoric often revolves around the appealing idea of eliminating so called red tape
        
**Challenge**
- to balance the need to protect the public with the competing need for economic growth.
- therein lies in political tale (controversy)
