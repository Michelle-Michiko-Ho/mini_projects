# Simulated Business Cycle Phases

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Time points
time = np.linspace(0, 20, 100)  # 0 to 20 time units

# GDP percent change (simulated data)
gdp_change = np.piecewise(
    time,
    [
        time < 5,    # Phase 1: Expansion
        (time >= 5) & (time < 10),  # Phase 2: Recession
        (time >= 10) & (time < 12),  # Phase 3: Trough
        (time >= 12) & (time < 15),  # Phase 4: Expansion
        (time >= 15) & (time < 20),  # Phase 5: Recession
        (time >= 20)  # Phase 6: Trough
    ],
    [
        lambda t: 2 * t - 1,           # Growth during expansion
        lambda t: -1 * (t - 5) + 2,    # Decline during recession
        lambda t: 1,                    # Low point at trough
        lambda t: 1.2 * (t - 2) + 1,     # Recovery during expansion
        lambda t: -0.5 * (t - 3) + 12,  # Decline during second recession
        lambda t: 5                   # Low point at second trough
    ]
)

# Reshape time for linear regression
time_reshaped = time.reshape(-1, 1)

# Fit a linear regression model
model = LinearRegression()
model.fit(time_reshaped, gdp_change)

# Predict values for the trend line
trend_line = model.predict(time_reshaped)

# Plotting the graph
plt.figure(figsize=(10, 5))
plt.plot(time, gdp_change, color='blue', linewidth=2, label='GDP Change')
plt.plot(time, trend_line, color='red', linestyle='--', linewidth=2, label='Trend Line')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')

# Adding labels and title
plt.title('Business Cycle Phases with Trend Line')
plt.xlabel('Time')
plt.ylabel('Real GDP Percent Change')
plt.ylim(-5, 30)
plt.xlim(0, 20)
plt.legend()

# Annotating the phases
plt.text(2.5, 0.5, 'Recovery', horizontalalignment='center', rotation=35)
plt.text(5, 9, 'Peak', horizontalalignment='center')
plt.text(7.5, 0.3, 'Recession', horizontalalignment='center', rotation=-20)
plt.text(10, -4, 'Trough', horizontalalignment='center')
plt.text(13.5, 11, 'Recovery', horizontalalignment='center', rotation=25)
plt.text(15, 17, 'Peak', horizontalalignment='center')
plt.text(17, 1.8, 'Recession', horizontalalignment='center', rotation=-10)
plt.text(19.5, 2, 'Trough', horizontalalignment='center')

# Show the plot
plt.grid()
plt.show()
```

# Recession Patterns:
-*like fingerprints, none is exactly identical*,

**Types**
- V-shape Pattern
   - short and deep
- U-shape Pattern
   - long and painful
     
**Double Dip Recession**
-*Right after recovery from a first recession, the economy dips into a second recession.*

- Nonrepeatable Internal Pattern
- Can't Predict Future Cycles with Historical Data

# Leading Economic Indicators:
-*used to measure Economic Indicators, as to see changes in advnace of underlying Business Cycle*
-*statistics or metrics that tend to change before the economy as a whole changes*


**Consumer Confidence Index**
- Rises
   - More confident
   - Consumers spend more
   - Consumpption up
   - GDP up
   - Bullish forecast
     
- Falls
   - Less confident
   - Consumers spend less
   - Consumption down
   - GDP down
   - Bearish forecast
  
 
**Types**

1. **Stock Market Performance**: Stock prices often reflect investor expectations about future corporate profits and economic conditions.

2. **Manufacturing Orders**: Increases in new orders for durable goods can signal future production increases, indicating economic growth.

3. **Building Permits**: The number of new building permits issued can indicate future construction activity and housing market trends.

4. **Consumer Confidence Index (CCI)**: Measures how optimistic or pessimistic consumers are about the economy, influencing their spending and saving behaviors.

5. **Retail Sales**: Changes in retail sales indicate consumer spending trends, which drive a significant portion of economic activity.

6. **Unemployment Claims**: A rise in initial jobless claims can signal a weakening job market, while a decline may indicate a strengthening economy.

7. **Average Weekly Hours Worked**: Increases in hours worked suggest greater demand for labor, which can be a precursor to hiring.

8. **Money Supply (M2)**: Changes in the money supply can influence interest rates and spending, affecting overall economic activity.

9. **Interest Rate Spreads**: The difference between short-term and long-term interest rates can indicate future economic conditions.

10. **Consumer Expectations**: Surveys measuring consumer expectations about inflation and economic conditions can provide insights into future spending.

These indicators are monitored closely by economists, businesses, and policymakers to gauge the likely direction of economic activity.

# Why Inflation Matters

**Rising Inflation Signaled by CPI**
- Central Bank Hikes Interest Rates
- Investments and Exports decrease
- GDP decreases
- Demand-pull inflation should subside

**Falling Inflation with slowing growth**
- Central Bank Lowers Interest Rates
- Investment and Exports increase
- GDP increases
