## <span style="color: dodgerblue;">Classifical Keynesian Debate in Modern Context:</span>

#### - Q1. How Active Should Gov't be
#### - Q2. Is It Gov't's Role to Implement Discretionary Policies to Solve MacroEcon Problems?
#### - Q3. Might Govn't Intervention Make the Situation Worse?


### 1. <span style="color: red;">Classical Price Adjustment Mechanism</span>
   - Prices Fall ==> Consumption Rises
   - Wages Fall  ==> Production Rises
   - Interest Rates Fall ==> Investment Rises

Leading to Economy returning to full employment.

**Main idea:**
- No gov't intervetion needed.
- Leave the market alone.
- *Laissez faire*

<span style="color: red;">Note: Favored by ideological conservatives now</span>

**Sum Up:**
- Recession is a natural part of the business cycle.
- Self-correcting
- No Govn't Intervention


### 2. <span style="color: blue;">Keynesian Income Adjustment Mechanism</span>
   - People's incomes fall (*in the absence of gov't intervention*)
        - Consumption falls ==> Business Production down
        - Savings fall  ==> Business Investment down

leading to deepending Recession

**Main Idea:**
- Discretionary Policy Intervention Necessary
- prescribe larger scale gov't expenditures to prime the conomic pump
- believe fiscal and monetary policies will positively affect the economy
- heavy intervention  

- thinkging...in the absence of gov't intervention...there may be a fall of
   - income
   - savings
   - investment
     
<span style="color: red;">Note: Favored by ideological liberals now</span>

# <li style="color: deepskyblue;font-weight:bold; font-size:30px">Say's law:</span>

- <span style="color: teal;">SUPPLY CREATES DEMAND</span>
- <span style="color: teal;">Create Supply Side: people work to produce (goods + services) ==> Earn Income </span>
- <span style="color: teal;">Create Demand Side: people spend all their income to buy all supply ==> Supply creates demand</span>

### &diams; <span style="color: blue;">Key Assumption: Income equals Value of Goods and Services produced </span> <a href="#classical"></a>


## <span style="color: brown;">Problem of Say's law:</span>

### Critiques:

- <span style="color: lightslategray;font style: bold;">By Thomas Malthus: People don't spend all their income</span>
  - <span style="color: green;font style: bold;">(Dark-sided view): Population will grow further than food production</span>
     - <span style="color: green;font style: bold;">Leading to an overpopulated starving world</span>
        - <span style="color: green;font style: bold;">Economics as the "Dismal Science"</span>
  - <span style="color: green;font style: bold;">(Logical view): Supply exceeds demand (glut of goods)</span>
     - <span style="color: green;font style: bold;">Leading to unemployment</span>

### Responses to the Critiques:

- <span style="color: lightslategray;font style: bold;">The money people don't spend will go to savings</span>
  - <span style="color: green;font style: bold;">Those savings will go to Banking / Finance / Business Investment</span>
     - <span style="color: green;font style: bold;">No unemployment</span>
        - <span style="color: green;font style: bold;">Aggregate Demand == Consumption Plus Investment == Aggregate Supply </span>
  - <span style="color: green;font style: bold;">Aggregate Supply</span>
     - <span style="color: green;font style: bold;">Employee Compensation, Rents, Interest, Profits</span>
  - <span style="color: green;font style: bold;">Aggregate Demand</span>
     - <span style="color: green;font style: bold;">Consumption, Investment</span>

# <li style="color: darkgray;font-weight:bold; font-size:30px;">Quantity Theory of Money:</span>

- <span style="color: teal;">aka Equation of Change</span>
- <span style="color: teal;">M * V = P * Q </span>
  - <span style="color: teal;">where M: money supply, V: velocity of money, P: price level, measured by index (CPI), Q: quantity level</span>
   - <span style="color: teal;">P * Q : Nominal GDP, where the changes of P determine Inflaction / Deflation</span>
    - <span style="color: teal;">P * Q : Real GDP, where Q is real inflation-adjusted output / Deflation</span>

### &diams; <span style="color: blue;">Key Concept: P varies in response to changes of Quantity of money </span> <a href="#classical"></a>
  - <span style="color: teal;">when M (money supply increases) ==> P (price of goods increases) ==> Inflation</span>
   - <span style="color: teal;">when M (money supply decreases) ==> P (price of goods decreases) ==> Deflation</span>

## <span style="color: brown;">How Effective is Monetary Policy?</span>
  - <span style="color: teal;">For instance, by increasing M will lead to increase real GDP growth Q?</span>
### Classical View of Monetary Policy:

- **<span style="color: lightslategray;">Print money to stimulate real GDP growth</span>**
  - **<span style="color: red;">Trigger Inflation, NOT effective at all!!!</span>**
     - <span style="color: green;font style: bold;">? Pessimistic view about Expansionary Monetary Policy</span>     

### &diams; <span style="color: blue;">Two Key Assumptions: </span> <a href="#classical"></a>
  - **<span style="color: teal;">1. Constant Velocity of Money</span>**
     - **<span style="color: brown;">Rate at which money is exchanged in a given time period</span>**
   - **<span style="color: teal;">2. Veil of Money</span>**
     - **<span style="color: brown;">Real money is not influnced by the real supply</span>**
        - **<span style="color: orange;">Govn't intervention using monetary policy is fruitless </span>**
            - **<span style="color: turquoise;">Money supply up while Real ouput Q unchanged</span>**


### implications:

- **<span style="color: green;">In Classical World, Supply always creates its demand - No need for Govn't Intervention<</span>**
  - **<span style="color: green;">The Economy will always find Employment Equilibrium</span>**
     - **<span style="color: green;">The only effect of an increase in the money supply beyond that needed tp sustain growth in real output would be INFLATION</span>**        
- **<span style="color: green;">Not Imply: An undersupply of money lead to a reduction in real output, resulting in recession or depression</span>**


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for real GDP
real_gdp = np.linspace(0, 100, 100)

# Adjusted Aggregate Demand (AD) and Aggregate Supply (AS) with steeper slopes
ad = 100 - 0.6 * real_gdp  # Steeper AD curve
as_supply = 30 + 0.8 * real_gdp  # Steeper AS curve

# Set the equilibrium point explicitly to (50, 70)
equilibrium_gdp = 50
equilibrium_price = 70

# Plotting the curves
plt.figure(figsize=(10, 6))
plt.plot(real_gdp, ad, label='Aggregate Demand (AD)', color='blue')
plt.plot(real_gdp, as_supply, label='Aggregate Supply (AS)', color='red')

# Mark the equilibrium point
plt.plot(equilibrium_gdp, equilibrium_price, 'go')  # Equilibrium point
plt.text(equilibrium_gdp, equilibrium_price, '  E', fontsize=12, verticalalignment='bottom', horizontalalignment='right')
plt.text(equilibrium_gdp, equilibrium_price + 5, f'P1: {equilibrium_price:.2f}', fontsize=10, verticalalignment='bottom', horizontalalignment='center')
plt.text(equilibrium_gdp + 5, equilibrium_price, f'Q1: {equilibrium_gdp:.2f}', fontsize=10, verticalalignment='center', horizontalalignment='left')

# Add dashed lines ending at the equilibrium point
plt.axvline(x=equilibrium_gdp, color='gray', linestyle='--', ymin=0, ymax=equilibrium_price/120)  # Vertical line at x=50
plt.axhline(y=equilibrium_price, color='gray', linestyle='--', xmin=0, xmax=equilibrium_gdp/100)  # Horizontal line at y=70
plt.text(51, 0, 'Q1', fontsize=10, verticalalignment='bottom', horizontalalignment='left')
plt.text(0, 71, 'P1', fontsize=10, verticalalignment='bottom', horizontalalignment='right')

# Add green solid vertical line at x=65
potential_output_x = 65
plt.axvline(x=potential_output_x, color='green', linestyle='-', label='Potential Output')  # Vertical line at x=65
plt.text(potential_output_x + 1, 10, 'Potential Output', fontsize=10, verticalalignment='bottom', horizontalalignment='left')
plt.text(potential_output_x, 0, 'Q2', fontsize=10, verticalalignment='bottom', horizontalalignment='left')

# Adding labels and title
plt.title('Aggregate Supply and Aggregate Demand')
plt.xlabel('Real GDP')
plt.ylabel('Price Level')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')

# Adding a legend
plt.legend()
plt.grid()

# Show the plot
plt.xlim(0, 100)
plt.ylim(0, 120)
plt.show()
```
