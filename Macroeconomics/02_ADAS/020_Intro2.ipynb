### <span style="color: red;"></span>

<span style="color: darkviolet; font-size:20px;font-weight:bold;">Why AD Curve Slopes Downward</span>


- ### **General Idea (Microeconomics)**
  - <span style="color: teal;">Due to the decrease of the price, consumers and businesses would increase their demand in goods and services</span>
  - <span style="color: teal;">The level of real GDP (domesetic output) will demand at each price level</span>

- ### **Macroeconomics Perspective**
*As a schedule of price quantity combinations, in each possible price level, holding others things constant, there is a desired quantity level correspondingly*
- <span style="color: teal;font-weight:bold;">Wealth Effect (Real Balance Effect)</span>
  - <span style="color: teal;">As the prices fall, the actual purchasing power increases</span>
- <span style="color: teal;font-weight:bold;">Interest Effect </span>
  - <span style="color: teal;">As the prices fall, interest rates typically fall </span>
    - <span style="color: teal;">Falling interest rates would increase investment spending </span>
    - <span style="color: teal;">More spending on interest-rate sensitive items (houses, cars)</span>
- <span style="color: teal;font-weight:bold;">Net Export Effect (Foreign Purchase / Trade Effect) </span>
  - <span style="color: teal;">If the Domestic prices fall... </span>
    - <span style="color: teal;">The domestic demand for foreign imports would increase or decrease? </span>
    - <span style="color: teal;">The domestic demand from foreigners for a country's exports increase or decrease?</span>
    - <span style="color: teal;">The Net Effect on Aggregate Demand</span>

### &diams; <span style="color: blue;">Caterer's Paribas Assumption: Holding Other Things Constant: </span></a>
  - **<span style="color: teal;">Isolates Specific Effects in the context of ADAS</span>**
     - **<span style="color: brown;">Why ADAS might shift inward / outward</span>**
     - **<span style="color: brown;">Effects represented by inward / ourward shifts in curves in two-dimentional space</span>**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for real GDP (Q)
Q = np.linspace(0, 100, 200)

# Define a steeper downward-sloping AD curve
# Adjust the slope (using -0.9 for steeper)
P_AD = 100 - 0.9 * Q

# Define price levels and corresponding GDP values
Q1, Q2, Q3 = 20, 50, 80
P1, P2, P3 = 100 - 0.9 * Q1, 100 - 0.9 * Q2, 100 - 0.9 * Q3

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(Q, P_AD, label='AD Curve', color='blue')
plt.scatter([Q1, Q2, Q3], [P1, P2, P3], color='red')

# Draw dashed lines to the axes
for Q_val, P_val in zip([Q1, Q2, Q3], [P1, P2, P3]):
    plt.axhline(y=P_val, xmin=0, xmax=Q_val/100, color='gray', linestyle='--')
    plt.axvline(x=Q_val, ymin=0, ymax=P_val/100, color='gray', linestyle='--')

# Annotate points
plt.text(Q1, P1, f'  P1: {P1:.1f}\n  Q1: {Q1}', fontsize=10, verticalalignment='bottom')
plt.text(Q2, P2, f'  P2: {P2:.1f}\n  Q2: {Q2}', fontsize=10, verticalalignment='bottom')
plt.text(Q3, P3, f'  P3: {P3:.1f}\n  Q3: {Q3}', fontsize=10, verticalalignment='bottom')

# Labels and title
plt.title('Why Aggregate Demand (AD) Curve is Downward Slopping\n', fontsize=14)
plt.text(0.5, 1.02, '(Moving Alongside Aggregate Demand Curve)', fontsize=10, ha='center', transform=plt.gca().transAxes)
plt.xlabel('Real GDP (Q)')
plt.ylabel('Price Level (P)')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```

<span style="color: darkgray;font-weight:bold;font-size:20px;">AD Moving Alongside vs Shifting In/Outward</span>

Essentially, the Aggregate Demand (AD) curve represents the total quantity of goods and services demanded across all levels of the economy at different price levels.

Understanding the difference between shifts in the AD curve (inward or outward) and movements along the curve is essential for analyzing economic changes. Hereâ€™s a breakdown of both concepts:

<span style="color: dodgerblue;font-weight:bold;font-size:20px;">AD Movements Along the Curve</span>

Movements along the AD curve occur in response to changes in the price level, leading to a change in the quantity of goods and services demanded without altering the underlying factors of aggregate demand.

**1. Movement Downward Along the AD Curve (Increase in Quantity Demanded):**  
This occurs when the price level falls, leading to an increase in the quantity of goods and services demanded.

**Example:** If the overall price level decreases due to deflation, consumers may increase their purchases because their purchasing power rises.

**2. Movement Upward Along the AD Curve (Decrease in Quantity Demanded):**  
This occurs when the price level rises, leading to a decrease in the quantity of goods and services demanded.

**Example:** If inflation occurs and prices rise, consumers may reduce their spending as their real purchasing power declines.

<span style="color: dodgerblue;font-weight:bold;font-size:20px;">AD Shifting In/Outward</span>

**1. Shifts Inward (Leftward Shift):**  
This indicates a decrease in aggregate demand, leading to lower output and prices at any given price level.

**Reasons for Inward Shifts:**
- **Decrease in Consumer Confidence:** If consumers are worried about the economy, they cut back on spending.
  - **Example:** A major financial crisis that leads to widespread job losses.

- **Reduction in Government Spending:** A cut in government programs or services can lead to a decrease in overall demand.
  - **Example:** A government decides to reduce its budget deficit, cutting public spending.

- **Monetary Policy Tightening:** Higher interest rates can reduce borrowing and spending.
  - **Example:** The central bank raises interest rates to combat inflation, leading to decreased consumer and business spending.

- **Decline in Exports:** A decrease in foreign demand for domestic goods reduces overall demand.
  - **Example:** A recession in key trading partner countries reduces their imports.

**2. Shifts Outward (Rightward Shift):**  
This indicates an increase in aggregate demand, leading to higher output and prices at any given price level. 

**Reasons for Outward Shifts:**
- **Increase in Consumer Confidence:** When consumers feel optimistic about their financial situation, they tend to spend more.
  - **Example:** A tax cut increases disposable income, prompting more consumer spending.
  
- **Government Spending Increases:** An increase in government expenditure on infrastructure or services boosts overall demand.
  - **Example:** A government stimulus package directed at building roads and schools.

- **Monetary Policy Easing:** Lower interest rates can encourage borrowing and spending.
  - **Example:** The central bank reduces interest rates, making loans cheaper for consumers and businesses.

- **Exports Increase:** A rise in foreign demand for domestic goods boosts overall demand.
  - **Example:** A global economic upturn increases demand for a countryâ€™s exports.




### Summary

- **Movements along the AD Curve** are solely the result of changes in the price level, affecting the quantity of goods and services demanded at those price levels.

- **Shifts in the AD Curve** (inward or outward) are caused by changes in economic factors like consumer confidence, government spending, monetary policy, and net exports. They represent a change in demand at all price levels.

Understanding these differences helps economists and policymakers gauge the health of an economy and implement appropriate responses.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the price level
P = np.linspace(0, 10, 100)

# Define the AD curves with a slope of -1
AD = 10 - P  # Original AD curve
AD_star = 12 - P  # Outward shift in AD curve

# Define the AS curves with a slope of 1
AS = P - 1.5  # Original AS curve (slope = 1.5)
AS_star = P + 0.5  # Outward shift in AS curve

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plotting AD curves
ax1.plot(AD, P, label='AD', color='blue')
ax1.plot(AD_star, P, label='AD*', color='orange', linestyle='--')
ax1.set_title('Aggregate Demand Curves')
ax1.set_xlabel('Quantity Demanded (AD)')
ax1.set_ylabel('Price Level (P)')

# Set the same limits for both plots
ax1.set_xlim(0, 15)  # Set x-axis limits
ax1.set_ylim(0, 10)  # Set y-axis limits

# Equilibrium points
q_eq = 6
p_eq_AD = 10 - q_eq  # For AD curve (should be 4)
p_eq_AD_star = 12 - q_eq  # For AD* curve (should be 6)

# Dashed lines for equilibrium points in AD plot
ax1.axvline(q_eq, color='gray', linestyle='--', lw=1)
ax1.axhline(p_eq_AD, color='gray', linestyle='--', lw=1)
ax1.axhline(p_eq_AD_star, color='gray', linestyle='--', lw=1)

# Labeling equilibrium points for AD
ax1.text(q_eq + 0.5, p_eq_AD, '(6, 4)', fontsize=10, color='blue')
ax1.text(q_eq + 0.5, p_eq_AD_star, '(6, 6)', fontsize=10, color='orange')

# Adding arrows for AD curves
ax1.annotate('', xy=(AD_star[20], P[20]), xytext=(AD[20], P[20]),
             arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
ax1.annotate('', xy=(AD_star[80], P[80]), xytext=(AD[80], P[80]),
             arrowprops=dict(arrowstyle='->', color='black', lw=1.5))

ax1.axhline(0, color='black', lw=0.5, ls='--')
ax1.axvline(0, color='black', lw=0.5, ls='--')
ax1.legend()
# ax1.grid()  # Remove this line to disable the grid

# Plotting AS curves
ax2.plot(AS, P, label='AS', color='green')
ax2.plot(AS_star, P, label='AS*', color='red', linestyle='--')
ax2.set_title('Aggregate Supply Curves')
ax2.set_xlabel('Quantity Supplied (AS)')
ax2.set_ylabel('Price Level (P)')

# Set the same limits for both plots
ax2.set_xlim(0, 15)  # Set x-axis limits
ax2.set_ylim(0, 10)  # Set y-axis limits

# Equilibrium points for AS
p_eq_AS = q_eq - 1.5  # For AS curve (should be 4.5)
p_eq_AS_star = q_eq + 0.5  # For AS* curve (should be 6.5)

# Dashed lines for equilibrium points in AS plot
ax2.axvline(q_eq, color='gray', linestyle='--', lw=1)
ax2.axhline(p_eq_AS, color='gray', linestyle='--', lw=1)
ax2.axhline(p_eq_AS_star, color='gray', linestyle='--', lw=1)

# Labeling equilibrium points for AS
ax2.text(q_eq + 0.5, p_eq_AS, '(6, 4.5)', fontsize=10, color='green')
ax2.text(q_eq + 0.5, p_eq_AS_star, '(6, 6.5)', fontsize=10, color='red')

# Adding arrows for AS curves
ax2.annotate('', xy=(AS_star[20], P[20]), xytext=(AS[20], P[20]),
             arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
ax2.annotate('', xy=(AS_star[80], P[80]), xytext=(AS[80], P[80]),
             arrowprops=dict(arrowstyle='->', color='black', lw=1.5))

ax2.axhline(0, color='black', lw=0.5, ls='--')
ax2.axvline(0, color='black', lw=0.5, ls='--')
ax2.legend()
# ax2.grid()  # Remove this line to disable the grid

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the price level
P = np.linspace(0, 10, 100)

# Define the AD curves
AD = 10 - P  # Original AD curve
AD_star = 12 - P  # Outward shift in AD curve

# Define the AS curves
AS = 1* P - 0.5  # Original AS curve
AS_star = 1 * P + 1.5  # Outward shift in AS curve

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plotting AD curves
ax1.plot(AD, P, label='AD', color='blue')
ax1.plot(AD_star, P, label='AD*', color='orange', linestyle='--')
ax1.set_title('Aggregate Demand Curves')
ax1.set_xlabel('Quantity Demanded (AD)')
ax1.set_ylabel('Price Level (P)')

# Adding arrows for AD curves
ax1.annotate('', xy=(AD_star[20], P[20]), xytext=(AD[20], P[20]),
             arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
ax1.annotate('', xy=(AD_star[80], P[80]), xytext=(AD[80], P[80]),
             arrowprops=dict(arrowstyle='->', color='black', lw=1.5))

# Adding one equilibrium point for AD and one for AD*
equilibrium_AD = (AD[20], P[20])  # E1 for AD
equilibrium_AD_star = (AD_star[40], P[40])  # E2 for AD*

# Plot and label equilibrium points
ax1.plot(*equilibrium_AD, 'ro')  
ax1.plot(*equilibrium_AD_star, 'ro')  
ax1.text(*equilibrium_AD, ' E1', fontsize=10, verticalalignment='bottom', horizontalalignment='right')
ax1.text(*equilibrium_AD_star, ' E2', fontsize=10, verticalalignment='bottom', horizontalalignment='right')

# Adding dashed lines for equilibrium points
ax1.axhline(y=equilibrium_AD[1], xmin=0, xmax=equilibrium_AD[0]/10, color='gray', linestyle='--')
ax1.axvline(x=equilibrium_AD[0], ymin=0, ymax=equilibrium_AD[1]/10, color='gray', linestyle='--')
ax1.axhline(y=equilibrium_AD_star[1], xmin=0, xmax=equilibrium_AD_star[0]/10, color='gray', linestyle='--')
ax1.axvline(x=equilibrium_AD_star[0], ymin=0, ymax=equilibrium_AD_star[1]/10, color='gray', linestyle='--')

ax1.axhline(0, color='black', lw=0.5, ls='--')
ax1.axvline(0, color='black', lw=0.5, ls='--')
ax1.legend()
ax1.grid()

# Plotting AS curves
ax2.plot(AS, P, label='AS', color='green')
ax2.plot(AS_star, P, label='AS*', color='red', linestyle='--')
ax2.set_title('Aggregate Supply Curves')
ax2.set_xlabel('Quantity Supplied (AS)')
ax2.set_ylabel('Price Level (P)')

# Adding arrows for AS curves
ax2.annotate('', xy=(AS_star[20], P[20]), xytext=(AS[20], P[20]),
             arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
ax2.annotate('', xy=(AS_star[80], P[80]), xytext=(AS[80], P[80]),
             arrowprops=dict(arrowstyle='->', color='black', lw=1.5))

# Adding one equilibrium point for AS and one for AS*
equilibrium_AS = (AS[15], P[15])  # E1 for AS
equilibrium_AS_star = (AS_star[35], P[35])  # E2 for AS*

# Plot and label equilibrium points
ax2.plot(*equilibrium_AS, 'ro')  
ax2.plot(*equilibrium_AS_star, 'ro')  
ax2.text(*equilibrium_AS, ' E1', fontsize=10, verticalalignment='bottom', horizontalalignment='right')
ax2.text(*equilibrium_AS_star, ' E2', fontsize=10, verticalalignment='bottom', horizontalalignment='right')

# Adding dashed lines for equilibrium points
ax2.axhline(y=equilibrium_AS[1], xmin=0, xmax=equilibrium_AS[0]/10, color='gray', linestyle='--')
ax2.axvline(x=equilibrium_AS[0], ymin=0, ymax=equilibrium_AS[1]/10, color='gray', linestyle='--')
ax2.axhline(y=equilibrium_AS_star[1], xmin=0, xmax=equilibrium_AS_star[0]/10, color='gray', linestyle='--')
ax2.axvline(x=equilibrium_AS_star[0], ymin=0, ymax=equilibrium_AS_star[1]/10, color='gray', linestyle='--')

ax2.axhline(0, color='black', lw=0.5, ls='--')
ax2.axvline(0, color='black', lw=0.5, ls='--')
ax2.legend()
ax2.grid()

# Show the plots
plt.tight_layout()
plt.show()
```
