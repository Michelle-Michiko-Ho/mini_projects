# Why AS Curve Slopes Upwawrd

- ### **General Idea (Microeconomics)**
  - <span style="color: teal;">Due to the increase of the price, businesses would increase their supply and sell add'l output as it provides incentive to do so.</span>
  - <span style="color: teal;">The level of real GDP (domesetic output) will produce at each price level</span> 


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for real GDP (Q)
Q = np.linspace(0, 100, 200)

# Define an upward-sloping AS curve
# Adjust the slope (using 0.9 for upward)
P_AS = 20 + 0.9 * Q

# Define price levels and corresponding GDP values for three points
Q1, Q2, Q3 = 20, 50, 80
P1, P2, P3 = 20 + 0.9 * Q1, 20 + 0.9 * Q2, 20 + 0.9 * Q3

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(Q, P_AS, label='AS Curve', color='green')
plt.scatter([Q1, Q2, Q3], [P1, P2, P3], color='red')

# Draw dashed lines to the axes
for Q_val, P_val in zip([Q1, Q2, Q3], [P1, P2, P3]):
    plt.axhline(y=P_val, xmin=0, xmax=Q_val/100, color='gray', linestyle='--')
    plt.axvline(x=Q_val, ymin=0, ymax=P_val/100, color='gray', linestyle='--')

# Annotate points
plt.text(Q1, P1, f'  P1: {P1:.1f}\n  Q1: {Q1}', fontsize=10, verticalalignment='top')
plt.text(Q2, P2, f'  P2: {P2:.1f}\n  Q2: {Q2}', fontsize=10, verticalalignment='top')
plt.text(Q3, P3, f'  P3: {P3:.1f}\n  Q3: {Q3}', fontsize=10, verticalalignment='top')

# Labels and title
plt.title('Why Aggregate Supply (AS) Curve is Upward Sloping\n', fontsize=14)
plt.text(0.5, 1.02, '(Moving Alongside Aggregate Supply Curve)', fontsize=10, ha='center', transform=plt.gca().transAxes)
plt.xlabel('Real GDP (Q)')
plt.ylabel('Price Level (P)')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
#plt.grid()
plt.legend()
plt.show()
```

Essentially, the Aggregate Demand (AD) curve represents the total quantity of goods and services demanded across all levels of the economy at different price levels.

Understanding the difference between shifts in the AD curve (inward or outward) and movements along the curve is essential for analyzing economic changes. Hereâ€™s a breakdown of both concepts:

### AD Shifts (Inward/Outward)

**1. Shifts Outward (Rightward Shift):**  
This indicates an increase in aggregate demand, leading to higher output and prices at any given price level. 

**Reasons for Outward Shifts:**
- **Increase in Consumer Confidence:** When consumers feel optimistic about their financial situation, they tend to spend more.
  - **Example:** A tax cut increases disposable income, prompting more consumer spending.
  
- **Government Spending Increases:** An increase in government expenditure on infrastructure or services boosts overall demand.
  - **Example:** A government stimulus package directed at building roads and schools.

- **Monetary Policy Easing:** Lower interest rates can encourage borrowing and spending.
  - **Example:** The central bank reduces interest rates, making loans cheaper for consumers and businesses.

- **Exports Increase:** A rise in foreign demand for domestic goods boosts overall demand.
  - **Example:** A global economic upturn increases demand for a countryâ€™s exports.

**2. Shifts Inward (Leftward Shift):**  
This indicates a decrease in aggregate demand, leading to lower output and prices at any given price level.

**Reasons for Inward Shifts:**
- **Decrease in Consumer Confidence:** If consumers are worried about the economy, they cut back on spending.
  - **Example:** A major financial crisis that leads to widespread job losses.

- **Reduction in Government Spending:** A cut in government programs or services can lead to a decrease in overall demand.
  - **Example:** A government decides to reduce its budget deficit, cutting public spending.

- **Monetary Policy Tightening:** Higher interest rates can reduce borrowing and spending.
  - **Example:** The central bank raises interest rates to combat inflation, leading to decreased consumer and business spending.

- **Decline in Exports:** A decrease in foreign demand for domestic goods reduces overall demand.
  - **Example:** A recession in key trading partner countries reduces their imports.

### AD Movements Along the Curve

Movements along the AD curve occur in response to changes in the price level, leading to a change in the quantity of goods and services demanded without altering the underlying factors of aggregate demand.

**1. Movement Downward Along the AD Curve (Increase in Quantity Demanded):**  
This occurs when the price level falls, leading to an increase in the quantity of goods and services demanded.

**Example:** If the overall price level decreases due to deflation, consumers may increase their purchases because their purchasing power rises.

**2. Movement Upward Along the AD Curve (Decrease in Quantity Demanded):**  
This occurs when the price level rises, leading to a decrease in the quantity of goods and services demanded.

**Example:** If inflation occurs and prices rise, consumers may reduce their spending as their real purchasing power declines.

### Summary

- **Shifts in the AD Curve** (inward or outward) are caused by changes in economic factors like consumer confidence, government spending, monetary policy, and net exports. They represent a change in demand at all price levels.
- **Movements along the AD Curve** are solely the result of changes in the price level, affecting the quantity of goods and services demanded at those price levels.

Understanding these differences helps economists and policymakers gauge the health of an economy and implement appropriate responses.

   - <span style="color: Fuchsia;">Which one is preferred as a way of seeing the real GDP rise  and why?</span>

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for real GDP (Q)
Q = np.linspace(0, 100, 200)

# Define AS and AD curves
P_AS = 20 + 0.5 * Q  # AS curve
P_AD = 100 - 0.9 * Q  # AD curve
P_AD_star = 120 - 0.9 * Q  # AD* curve (shifted further outward)
P_AS_star = 5 + 0.5 * Q  # AS* curve (shifted further downward)

# Solve for the original equilibrium point
Q_eq = 80 / 1.4
P_eq = 20 + 0.5 * Q_eq  # Original equilibrium price

# Solve for the new equilibrium point with AD*
Q_eq_star = 100 / 1.4
P_eq_star = 20 + 0.5 * Q_eq_star  # New equilibrium price with AD*

# Solve for the new equilibrium point with AS*
Q_eq_star_AS = 95 / 1.4
P_eq_star_AS = 5 + 0.5 * Q_eq_star_AS  # New equilibrium price with AS*

# Create the plot with two subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Plotting in the left graph (AD and AD*)
axs[0].plot(Q, P_AS, label='AS Curve', color='green')
axs[0].plot(Q, P_AD, label='AD Curve', color='blue')
axs[0].plot(Q, P_AD_star, label='AD* Curve', color='orange')

# Plot the original equilibrium point
axs[0].scatter(Q_eq, P_eq, color='red', label='Equilibrium Point (AD vs AS)')  # Mark original equilibrium point
axs[0].text(Q_eq, P_eq, f'  P: {P_eq:.1f}\n  Q: {Q_eq:.1f}', fontsize=10, verticalalignment='bottom', color='red')

# Draw dashed lines for the original equilibrium point
axs[0].axhline(y=P_eq, xmin=0, xmax=Q_eq/100, color='gray', linestyle='--')
axs[0].axvline(x=Q_eq, ymin=0, ymax=P_eq/100, color='gray', linestyle='--')

# Plot the new equilibrium point
axs[0].scatter(Q_eq_star, P_eq_star, color='purple', label='Equilibrium Point (AD* vs AS)')  # Mark new equilibrium point
axs[0].text(Q_eq_star, P_eq_star, f'  P*: {P_eq_star:.1f}\n  Q*: {Q_eq_star:.1f}', fontsize=10, verticalalignment='bottom', color='purple')

# Draw dashed lines for the new equilibrium point
axs[0].axhline(y=P_eq_star, xmin=0, xmax=Q_eq_star/100, color='gray', linestyle='--')
axs[0].axvline(x=Q_eq_star, ymin=0, ymax=P_eq_star/100, color='gray', linestyle='--')

# Set labels and title for the left subplot
axs[0].set_xlabel('Real GDP (Q)', color='black', fontsize=12)
axs[0].set_ylabel('Price Level (P)', color='black', fontsize=12)
axs[0].set_xlim(0, 100)
axs[0].set_ylim(0, 100)
axs[0].axhline(0, color='black', linewidth=0.5, ls='--')
axs[0].axvline(0, color='black', linewidth=0.5, ls='--')
#axs[0].grid()
axs[0].legend()

# Add labels for equilibrium points on axes with colors and larger font size
axs[0].text(Q_eq, -5, 'Q', fontsize=14, horizontalalignment='center', color='red')
axs[0].text(-5, P_eq, 'P', fontsize=14, verticalalignment='center', rotation=00, color='red')
axs[0].text(Q_eq_star, -5, 'Q*', fontsize=14, horizontalalignment='center', color='purple')
axs[0].text(-5, P_eq_star, 'P*', fontsize=14, verticalalignment='center', rotation=0, color='purple')

# Plotting in the right graph (AD and AS*)
axs[1].plot(Q, P_AS, label='AS Curve', color='green')
axs[1].plot(Q, P_AD, label='AD Curve', color='blue')
axs[1].plot(Q, P_AS_star, label='AS* Curve', color='orange')

# Plot the original equilibrium point
axs[1].scatter(Q_eq, P_eq, color='red', label='Equilibrium Point')  # Mark equilibrium point
axs[1].text(Q_eq, P_eq, f'  P: {P_eq:.1f}\n  Q: {Q_eq:.1f}', fontsize=10, verticalalignment='bottom', color='red')

# Draw dashed lines to axes from the equilibrium point
axs[1].axhline(y=P_eq, xmin=0, xmax=Q_eq/100, color='gray', linestyle='--')
axs[1].axvline(x=Q_eq, ymin=0, ymax=P_eq/100, color='gray', linestyle='--')

# Plot the new equilibrium point with AS*
axs[1].scatter(Q_eq_star_AS, P_eq_star_AS, color='purple', label='Equilibrium Point (AD vs AS*)')  # Mark new equilibrium point
axs[1].text(Q_eq_star_AS, P_eq_star_AS, f'  P*: {P_eq_star_AS:.1f}\n  Q*: {Q_eq_star_AS:.1f}', fontsize=10, verticalalignment='bottom', color='purple')

# Draw dashed lines for the new equilibrium point
axs[1].axhline(y=P_eq_star_AS, xmin=0, xmax=Q_eq_star_AS/100, color='gray', linestyle='--')
axs[1].axvline(x=Q_eq_star_AS, ymin=0, ymax=P_eq_star_AS/100, color='gray', linestyle='--')

# Set labels and title for the right subplot
axs[1].set_xlabel('Real GDP (Q)', color='black', fontsize=12)
axs[1].set_ylabel('Price Level (P)', color='black', fontsize=12)
axs[1].set_xlim(0, 100)
axs[1].set_ylim(0, 100)
axs[1].axhline(0, color='black', linewidth=0.5, ls='--')
axs[1].axvline(0, color='black', linewidth=0.5, ls='--')
#axs[1].grid()
axs[1].legend()

# Add labels for equilibrium points on axes in the right plot with larger font size
axs[1].text(Q_eq, -5, 'Q', fontsize=14, horizontalalignment='center', color='red')
axs[1].text(-5, P_eq, 'P', fontsize=14, verticalalignment='center', rotation=0, color='red')
axs[1].text(Q_eq_star_AS, -5, 'Q*', fontsize=14, horizontalalignment='center', color='purple')
axs[1].text(-5, P_eq_star_AS, 'P*', fontsize=14, verticalalignment='center', rotation=90, color='purple')

# Set the title for the overall figure
plt.suptitle('Aggregate Demand and Supply Curves', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
def draw_oval_outline(rows, cols):
    for i in range(rows):
        for ja in range(cols):
            # Calculate the distance from the center to create an oval outline
            distance = ((i - rows // 2) ** 2) / (rows // 2) ** 2 + ((j - cols // 2) ** 2) / (cols // 2) ** 2
            
            # Print arrows to form a continuous oval shape
            if abs(distance - 1) < 0.02:  # Threshold for oval outline
                if i == 0 and j % 2 == 0:  # Top edge
                    print('â†’', end=' ')
                elif i == rows - 1 and j % 2 == 0:  # Bottom edge
                    print('â†', end=' ')
                elif j == 0:  # Left edge
                    print('â†‘', end=' ')
                elif j == cols - 1:  # Right edge
                    print('â†“', end=' ')
                else:
                    print('â†’', end=' ')
            else:
                print(' ', end=' ')
        print()

# Set the dimensions of the oval
draw_oval_outline(15, 40)
```
