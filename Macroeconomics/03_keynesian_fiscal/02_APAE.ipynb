```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

# Function to update the plot
def update_plot(AEe=1500, Q_line=600):
    # Clear the previous plot
    plt.clf()

    # Real GDP (Output) range
    GDP = np.linspace(0, 1200, 100)

    # Calculate the equilibrium point Qe from AEe
    # AE = 0.8 * GDP + b
    # At equilibrium, AE = Y = X ==> AEe = Qe
    Qe = AEe  # At equilibrium, AE = Qe

    # The slope of AE is calculated as m = (AEe - 200) / Qe
    m = (AEe - 200) / Qe  # The slope of AE changes dynamically with AEe

    # The intercept b is then calculated from AE = 0.8 * Qe + b => b = AEe - m * Qe
    b = AEe - m * Qe

    # Aggregate Expenditures function
    AE = m * GDP + b

    # Aggregate Production (the 45-degree line)
    production = GDP  # Y = X line, representing Aggregate Production

    # Plotting the graph
    plt.figure(figsize=(8, 6))

    # Plot the Aggregate Expenditures (AE) line
    plt.plot(GDP, AE, label="Aggregate Expenditures (AE = C + I + G + X)", color='blue', linewidth=2)

    # Plot the Aggregate Production (45-degree line)
    plt.plot(GDP, production, label="Aggregate Production (Y = X)", color='red', linestyle='--', linewidth=2)

    # Plot the equilibrium point
    plt.scatter(Qe, AEe, color='green', zorder=5)  # Equilibrium point (Qe, AEe)
    plt.text(Qe + 30, AEe - 30, f'Equilibrium\n(Qe={Qe:.2f}, AEe={AEe})', color='green', fontsize=12)

    # Draw the vertical line from the equilibrium point to the y-axis (AEe line)
    #plt.plot([Qe, Qe], [0, AEe], color='green', linestyle=':', label="AEe Line")
    plt.plot([Qe, Qe], [0, AEe], color='green', linestyle=':')
    
    # Draw the horizontal line from the equilibrium point to the x-axis (Qe line)
    #plt.plot([0, Qe], [AEe, AEe], color='green', linestyle=':', label="Qe Line")
    plt.plot([0, Qe], [AEe, AEe], color='green', linestyle=':')

    # Plot the vertical line based on Q_line slider value
    plt.axvline(x=Q_line, color='purple', linestyle='-', linewidth=2, label=f"Production Potential")

    # Set the limits of the axes so the green lines touch the axes
    plt.xlim(0, max(Qe, Q_line) + 100)  # Extend the x-limits based on Qe and Q_line
    plt.ylim(0, max(AEe, Q_line) + 100)  # Extend the y-limits based on AEe and Q_line

    # Adding labels and title
    plt.title('Aggregate Expenditures vs. Real GDP Output', fontsize=14)
    plt.xlabel('Real GDP Output', fontsize=12)
    plt.ylabel('Aggregate Expenditures (AE)', fontsize=12)

    # Displaying the legend and grid
    plt.legend(loc="upper left")
    #plt.grid(True)

    # Show the plot
    plt.show()

# Create a slider for AEe value adjustment
slider_AEe = widgets.FloatSlider(
    value=1000,  # Initial value of AEe
    min=0,       # Minimum value
    max=1500,    # Maximum value
    step=10,     # Step size
    description='AE Level:',
    continuous_update=False
)

# Create a slider for Q_line value adjustment
slider_Q_line = widgets.FloatSlider(
    value=600,  # Initial value of Q_line
    min=0,      # Minimum value
    max=1200,   # Maximum value
    step=10,    # Step size
    description='Production Potential (Q Line):',
    continuous_update=False
)

# Create an interactive plot with both sliders
interact(update_plot, AEe=slider_AEe, Q_line=slider_Q_line)
```

The assumption that **prices are fixed** in certain Keynesian models is central to the **Keynesian cross** and other short-run models of aggregate demand. This assumption of "price rigidity" or "sticky prices" is particularly relevant in the context of Keynes's own *General Theory of Employment, Interest, and Money* (1936), and it was further developed and popularized by economists like Alvin Hansen and Paul Samuelson. Here's a breakdown of where this assumption comes from in their work:

### **Keynes's Original Model:**
- **John Maynard Keynes** was the first to introduce the assumption of fixed prices in his *General Theory* (1936). In the short run, Keynes assumed that prices and wages are **rigid** or "sticky," meaning they do not adjust immediately in response to changes in demand or supply.
  - This rigidity was an important feature of Keynes's model because it explained why economies could experience prolonged periods of unemployment or underemployment equilibrium. In a **classical** model, prices would adjust quickly to restore full employment, but Keynes argued that in the real world, wages and prices often don't adjust quickly enough to clear the labor and goods markets. This led to situations where demand deficiencies could cause persistent unemployment, something that classical economics could not explain.
  
  - **Keynesian Model Assumption of Fixed Prices:**
    - Keynes assumed that in the short run, **prices were fixed** (or sticky) and could not easily adjust to changes in aggregate demand. In this model, the economy could be in equilibrium with unemployment and output below full capacity, because prices (including wages) did not adjust sufficiently to restore full employment.
  
    - This assumption was crucial because it allowed Keynes to argue that **government intervention**, through fiscal policy (e.g., increasing public spending), was necessary to boost aggregate demand and reduce unemployment.

### **Hansen’s Development of the Keynesian Model:**
- **Alvin Hansen**, a prominent Keynesian economist, built on Keynes's framework, especially in the U.S. context during the Great Depression and the post-World War II era.
  
  - In his writings, including his key texts like *"Full Employment or the New Economics"* (1941) and *"The Economic Problem of the American Worker"* (1940), Hansen incorporated the idea of **fixed prices** in his models of aggregate demand and supply.
  
  - **Hansen's Fixed Price Assumption:**
    - Hansen used the Keynesian model to argue that in the short run, price flexibility was limited, and therefore, economies could experience equilibrium at less than full employment. This was a core reason why he supported **active government intervention** to stimulate demand, as the economy could not self-correct quickly through price adjustments alone.
    - He explained that the primary mechanism for restoring equilibrium was not through price changes, but through **changes in aggregate demand** driven by government spending, investment, and consumption.
  
  - In the context of *secular stagnation* theory, Hansen also suggested that even when prices were fixed, there was a risk of prolonged stagnation because insufficient demand would fail to encourage the needed investment.

### **Samuelson's Keynesian Synthesis:**
- **Paul Samuelson**, in his textbook *Economics* (first published in 1948), played a crucial role in popularizing the Keynesian model in the U.S. and introducing it to a wider audience. In his interpretation of Keynesian economics, Samuelson used the **fixed prices assumption** in much the same way Keynes did in the short-run macroeconomic models.
  
  - **Samuelson’s Key Contribution:**
    - Samuelson's version of the Keynesian model, especially in the form of the **Keynesian cross** or the aggregate demand-aggregate supply (AD-AS) framework, assumed that **prices were fixed** in the short run. This allowed for a model where **aggregate demand** could determine the level of output and employment.
    - In his model, **output** was determined by the level of **aggregate demand** (which consisted of consumption, investment, and government spending), and prices were assumed to be fixed in the short run because of wage and price rigidity. The economy could end up in an equilibrium where output was below full employment, and prices would not adjust to restore full employment automatically.
  
    - The **Keynesian Cross Model** that Samuelson and others popularized is one of the simplest forms of the Keynesian model, and it explicitly assumes that prices are fixed, at least in the short run. In this model, the equilibrium level of output is determined by the point where the **aggregate demand curve intersects the 45-degree line**, which represents all points where output equals income. The focus is on the demand side rather than the supply side.

### **Specific References to Fixed Prices in Hansen and Samuelson's Works:**

1. **Alvin Hansen's "Full Employment or the New Economics" (1941):**
   - In this book, Hansen develops Keynesian ideas and discusses the role of government in achieving full employment. He frequently refers to the **rigidity of prices** in explaining why demand deficiencies could lead to unemployment and why policy intervention was necessary. 
   - For example, he writes about how during the Depression, wages and prices were **sticky**, meaning that simply allowing wages and prices to adjust downward would not necessarily restore full employment.

2. **Paul Samuelson's "Economics" (1948) and Subsequent Editions:**
   - Samuelson's textbook is perhaps the most well-known place where the assumption of fixed prices in the Keynesian model is clearly explained. In the early chapters of *Economics*, Samuelson discusses Keynesian economics and the concept of the Keynesian cross. He explains that in the short run, **prices are fixed** in this model, and equilibrium is determined by aggregate demand rather than the interaction of supply and demand as in classical models.
   - Later editions of the textbook (e.g., in the 1960s and 1970s) continued to emphasize this fixed-price assumption in the analysis of short-run macroeconomic dynamics, especially in the context of the business cycle.

3. **Samuelson's "Foundations of Economic Analysis" (1947):**
   - In his earlier work, *Foundations of Economic Analysis*, Samuelson used mathematical models to present Keynesian ideas. While this book is more abstract and technical, it laid out formal models of aggregate demand and aggregate supply where price rigidity was assumed for simplicity and clarity. 

### **Why the Fixed Price Assumption?**
   - **Price Rigidity:** Both Hansen and Samuelson, like Keynes, assumed that prices (including wages) did not adjust instantly in the short run due to factors like **labor contracts**, **monopolistic practices**, or **menu costs** (the cost to firms of changing prices). This was crucial for explaining why the economy could settle into an equilibrium with unemployment or underemployment without the immediate self-correction that classical models predicted.
   - **Focus on Aggregate Demand:** By assuming fixed prices, Keynesian models could focus on **aggregate demand** as the main determinant of output and employment, without needing to complicate the model with price and wage adjustments. The idea was that fiscal policy could boost demand to reduce unemployment when the economy was operating below its potential output.

### **Conclusion:**
Both **Alvin Hansen** and **Paul Samuelson** built upon Keynes’s original assumption of fixed prices (price rigidity) in their versions of the Keynesian model. This assumption allowed them to focus on aggregate demand as the main determinant of output and employment, making it clear that **government intervention** was necessary to boost demand and achieve full employment. While this assumption has been modified and refined over time, it remains a cornerstone of short-run Keynesian economics, particularly when explaining the persistence of unemployment and economic downturns.
