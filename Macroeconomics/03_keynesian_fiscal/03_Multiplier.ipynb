```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

# Function to update the plot
def update_plot(AEe=1500, Q_line=600):
    # Clear the previous plot
    plt.clf()

    # Real GDP (Output) range
    GDP = np.linspace(0, 1200, 100)

    # Calculate the equilibrium point Qe from AEe
    # AE = 0.8 * GDP + b
    # At equilibrium, AE = Y = X ==> AEe = Qe
    Qe = AEe  # At equilibrium, AE = Qe

    # The slope of AE is calculated as m = (AEe - 200) / Qe
    m = (AEe - 200) / Qe  # The slope of AE changes dynamically with AEe

    # The intercept b is then calculated from AE = 0.8 * Qe + b => b = AEe - m * Qe
    b = AEe - m * Qe

    # Aggregate Expenditures function
    AE = m * GDP + b

    # Aggregate Production (the 45-degree line)
    production = GDP  # Y = X line, representing Aggregate Production

    # Plotting the graph
    plt.figure(figsize=(8, 6))

    # Plot the Aggregate Expenditures (AE) line
    plt.plot(GDP, AE, label="Aggregate Expenditures (AE = C + I + G + X)", color='blue', linewidth=2)

    # Plot the Aggregate Production (45-degree line)
    plt.plot(GDP, production, label="Aggregate Production (Y = X)", color='red', linestyle='--', linewidth=2)

    # Plot the equilibrium point
    plt.scatter(Qe, AEe, color='green', zorder=5)  # Equilibrium point (Qe, AEe)
    plt.text(Qe + 30, AEe - 30, f'Equilibrium\n(Qe={Qe:.2f}, AEe={AEe})', color='green', fontsize=12)

    # Draw the vertical line from the equilibrium point to the y-axis (AEe line)
    plt.plot([Qe, Qe], [0, AEe], color='green', linestyle=':', label="AEe Line")
    
    # Draw the horizontal line from the equilibrium point to the x-axis (Qe line)
    plt.plot([0, Qe], [AEe, AEe], color='green', linestyle=':', label="Qe Line")

    # Plot the vertical line based on Q_line slider value
    plt.axvline(x=Q_line, color='purple', linestyle='-', linewidth=2, label=f"Production Potential")

    # Set the limits of the axes so the green lines touch the axes
    plt.xlim(0, max(Qe, Q_line) + 100)  # Extend the x-limits based on Qe and Q_line
    plt.ylim(0, max(AEe, Q_line) + 100)  # Extend the y-limits based on AEe and Q_line

    # Adding labels and title
    plt.title('Aggregate Expenditures vs. Real GDP Output', fontsize=14)
    plt.xlabel('Real GDP Output', fontsize=12)
    plt.ylabel('Aggregate Expenditures (AE)', fontsize=12)

    # Displaying the legend and grid
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

# Create a slider for AEe value adjustment
slider_AEe = widgets.FloatSlider(
    value=1000,  # Initial value of AEe
    min=0,       # Minimum value
    max=1500,    # Maximum value
    step=10,     # Step size
    description='AE Level:',
    continuous_update=False
)

# Create a slider for Q_line value adjustment
slider_Q_line = widgets.FloatSlider(
    value=600,  # Initial value of Q_line
    min=0,      # Minimum value
    max=1200,   # Maximum value
    step=10,    # Step size
    description='Production Potential (Q Line):',
    continuous_update=False
)

# Create an interactive plot with both sliders
interact(update_plot, AEe=slider_AEe, Q_line=slider_Q_line)
```

<span style="color: darkturquoise; font-size: 30px; font-weight: bold;">Great Depression: 1929</span>


The Great Depression of 1929 was triggered by a combination of factors, many of which had been building over the preceding years. The immediate cause was the **stock market crash** in October 1929, but this event was part of a broader set of economic problems that made the economy vulnerable. Hereâ€™s a summary of the key factors that contributed to the onset of the Great Depression:

### 1. **The Stock Market Crash of 1929**
   - **Black Thursday (October 24, 1929)** and **Black Tuesday (October 29, 1929)** were the most famous days of the stock market crash. Stock prices had been soaring through the 1920s, fueled by speculation. Many people were investing heavily in stocks using borrowed money (on margin), assuming prices would keep rising.
   - When stock prices began to fall, panic selling ensued. This created a **vicious cycle**: falling stock prices led to more margin calls (requiring investors to repay their loans), which led to further sales and deeper declines.
   - By the end of 1929, the stock market had lost nearly **90%** of its value.

### 2. **Overproduction and Underconsumption**
   - During the 1920s, businesses, particularly in agriculture and manufacturing, were producing more goods than consumers could buy. This was partly due to **technological advances** that allowed for increased production, but wages for workers were not rising fast enough to keep up with the output.
   - As a result, companies had **unsold inventory**, leading to layoffs and further reducing purchasing power in the economy. This created a **downward spiral** of reduced demand, falling wages, and increased unemployment.

### 3. **Bank Failures and Financial Instability**
   - Many banks had invested heavily in the stock market, and when the market collapsed, they suffered huge losses. As a result, **thousands of banks failed** in the early years of the Depression.
   - People panicked and rushed to withdraw their money, causing a **banking crisis**. By 1933, nearly **9,000 banks** had failed, which further eroded public confidence in the financial system.
   - The lack of a central banking system to regulate monetary policy also contributed to the crisis. The Federal Reserve failed to act swiftly to prevent the crisis from spreading.

### 4. **Agricultural Collapse**
   - The **1920s was a difficult time for farmers**. The demand for agricultural products had fallen after World War I, and many farmers were left with massive debt.
   - In addition, the **Dust Bowl** (a series of severe droughts in the mid-1930s) devastated crops in the Great Plains, particularly in states like Oklahoma and Kansas. The agricultural sector, which was already struggling, was further harmed by environmental factors and economic mismanagement.

### 5. **Unequal Distribution of Wealth**
   - Income inequality was a significant issue during the 1920s. A large portion of the wealth was concentrated in the hands of a small elite, while many Americans (especially workers) had little purchasing power.
   - This disparity meant that even though the economy was producing a lot, the majority of people couldn't afford to buy the goods being produced, leading to **underconsumption**. When the economic downturn hit, the lack of a broad consumer base to sustain demand made recovery difficult.

### 6. **International Economic Factors**
   - **The Smoot-Hawley Tariff (1930)**, passed by the U.S. Congress, raised tariffs on foreign goods in an effort to protect American industries. However, this backfired by causing other countries to retaliate with tariffs on American goods, leading to a **decline in international trade**.
   - Many European countries were still recovering from the effects of World War I, and the global economy was weak. The U.S. also had invested heavily in European nations, and when the Depression hit, European markets collapsed, dragging down international trade further.

### 7. **Monetary Policy and the Gold Standard**
   - The **Gold Standard** (which fixed the value of money to a specific amount of gold) also played a role. As the U.S. was on the Gold Standard, the Federal Reserve was limited in its ability to adjust the money supply to respond to the crisis. Instead of increasing liquidity, the Fed **raised interest rates** in an attempt to protect the value of the dollar, which worsened the crisis by contracting the money supply further.
   - Many countries also remained on the Gold Standard, which meant they could not devalue their currencies or use fiscal policy effectively to counter the downturn.

### 8. **Weak Government Response**
   - Initially, President **Herbert Hoover** believed that the market would recover on its own. His response to the crisis was generally seen as insufficient and ineffective. Hoover resisted large-scale government intervention and believed that local communities and businesses should solve the problems.
   - However, as the crisis deepened, Hoover eventually attempted some measures like public works projects and government loans to businesses, but these were too little, too late to reverse the economic collapse.

### Summary: The Perfect Storm
The Great Depression was caused by a **combination of economic mismanagement**, **financial instability**, **speculative excess**, and **global interconnectedness**. The stock market crash was the immediate trigger, but the underlying causes were structural problems that had been building up for years. The collapse of the financial system, coupled with agricultural and industrial overproduction, weak government responses, and international factors, led to an economic collapse that lasted for much of the 1930s.

In response, the U.S. government, under **Franklin D. Roosevelt**, enacted a series of programs known as the **New Deal** to address the economic challenges, stabilize the banking system, and provide relief to the unemployed. The effects of the Depression were felt worldwide, and it took World War II and the wartime economy to finally pull many countries out of the depression.

<span style="color: dodgerblue; font-size: 30px; font-weight: bold;">Banking Crisis: 2023</span>

The banking crisis that unfolded in 2023, particularly with high-profile failures like **Silicon Valley Bank (SVB)** and **Signature Bank**, was caused by a combination of factors. These included rising interest rates, poor risk management, and a lack of confidence in the stability of some financial institutions. Here's a breakdown of the key factors that triggered the 2023 banking crisis:

### 1. **Rising Interest Rates and Bond Losses**
   - The **Federal Reserve's aggressive interest rate hikes** throughout 2022 and 2023 were a central factor. In response to rising inflation, the Fed raised rates at a pace not seen in decades. The goal was to cool down the economy and bring inflation back to the target range.
   - However, **higher interest rates** caused the value of long-term **bonds** (like U.S. Treasury bonds and mortgage-backed securities) to fall. Many banks, including SVB, had significant holdings in long-duration bonds. These assets lose value when interest rates rise, leading to **paper losses** in their portfolios.
   - **Silicon Valley Bank**, for example, had a large proportion of its portfolio tied to U.S. government bonds, which it had bought when interest rates were low. When rates rose, the market value of these bonds dropped, leaving the bank with unrealized losses.

### 2. **Unrealized Losses and Liquidity Problems**
   - As interest rates increased, many banks, including SVB, were sitting on **unrealized losses** on their bond portfolios. These losses were not immediately reflected in the banksâ€™ financial statements, but they were a significant vulnerability.
   - **SVB**'s client base was heavily concentrated in the **tech sector** and startups, which were facing a **funding squeeze** as venture capital funding slowed down due to higher borrowing costs. This led to a surge in **withdrawals** from SVB, as startups needed cash to cover expenses, and venture capital firms pulled funds to keep their operations going.
   - To meet withdrawal demands, **SVB was forced to sell its bonds** at a loss, realizing the negative impact of rising interest rates. The realization of these losses caused concerns among its depositors and led to a **bank run**.

### 3. **Concentration of Clients and Lack of Diversification**
   - **Silicon Valley Bank** had a very **concentrated client base**, with many of its depositors being **tech companies, startups, and venture capitalists**. When these clients started withdrawing funds due to the liquidity crunch in the tech sector, it had a **disproportionate impact** on the bank's stability.
   - **Signature Bank**, similarly, had a high concentration of deposits from the **cryptocurrency sector**, which was also facing volatility in 2023. As the crypto market struggled, Signature Bank saw a large portion of its deposits withdrawn.
   - The lack of **diversification** in client bases and in assets made these banks more vulnerable to sector-specific downturns and to macroeconomic pressures.

### 4. **Lack of Confidence and Bank Runs**
   - The **collapse of SVB** triggered a broader loss of confidence in the banking sector. As news of SVB's troubles spread, depositors at other banks began to worry about their own financial institutions' stability.
   - **Bank runs** can happen when depositors lose faith in a bankâ€™s solvency and rush to withdraw their funds. Even though most banks are insured up to $250,000 by the **FDIC**, the fear of insolvency and loss of access to funds led to a rapid **exodus of deposits** from smaller banks.
   - **Signature Bank** and other smaller banks faced similar runs in the aftermath of SVB's failure.

Fiscal policy can "crowd out" private investment when government spending or borrowing increases to the point that it negatively affects private sector investment. This typically occurs through the mechanism of rising interest rates. Here's a detailed explanation of how this happens:

### 1. **Increased Government Borrowing:**
   When the government increases its spending, especially in times of budget deficits, it often borrows money to finance the difference between its revenue and its expenditures. This is typically done by issuing government bonds.

   - **Government borrowing increases demand for loanable funds:** As the government borrows more, it competes with private borrowers in the financial markets for available funds.
   - **Demand for funds increases:** With more borrowers in the market (the government being a major one), the total demand for money (or loanable funds) rises.

### 2. **Rising Interest Rates:**
   The increased demand for funds (due to higher government borrowing) tends to drive up interest rates, assuming the supply of loanable funds doesn't increase correspondingly.

   - **Interest rates rise:** As demand for money increases, lenders can charge higher interest rates because they have more borrowers competing for the same pool of funds.
   
### 3. **Higher Cost of Borrowing for Private Sector:**
   When interest rates rise, it becomes more expensive for private businesses to borrow money to fund their own investments, such as purchasing new equipment, expanding operations, or undertaking research and development.

   - **Private investment becomes less attractive:** Higher interest rates increase the cost of capital for businesses. In response, businesses may reduce or postpone their investment plans, as the return on investment (ROI) needed to justify borrowing becomes harder to achieve.

### 4. **Crowding Out:**
   This reduction in private investment due to the higher cost of borrowing is referred to as "crowding out." Essentially, the government's increased borrowing reduces the funds available for private businesses to invest, leading to a reduction in overall private investment.

### 5. **Possible Long-Term Effects:**
   Over time, if government spending remains high and borrowing continues to increase, the crowding-out effect could lead to a sustained reduction in private sector investment. This might slow down economic growth, as private investment is a key driver of productivity, innovation, and job creation.

### Example:

Imagine that a government launches a large infrastructure program, and to finance it, it borrows heavily by issuing bonds. The demand for loanable funds rises because the government is absorbing a large portion of the available funds. This increases interest rates, making it more expensive for a tech startup, for instance, to get a loan to expand its operations. As a result, the startup might decide to delay its expansion, leading to a reduction in private sector investment.

### How to Mitigate Crowding Out:

- **Monetary Policy Response:** The central bank could lower interest rates (e.g., through open market operations) to offset the increase in rates caused by government borrowing, which could help stimulate private investment even in the face of higher government spending.
- **Well-Targeted Government Spending:** If government spending is directed toward projects that enhance productivity (like education, infrastructure, or research), the economy as a whole may benefit, making it easier for private investment to thrive alongside public investment.

### Conclusion:
Fiscal policy crowds out private investment mainly through higher government borrowing, which increases interest rates, making it more expensive for businesses to invest. While this is a typical concern in economies with high levels of government debt, the overall impact on private investment depends on the size of the fiscal expansion, the responsiveness of monetary policy, and the state of the economy.
